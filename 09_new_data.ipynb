```python
import pandas as pd
import numpy as np

df = pd.read_csv("panel_ML_df.csv")
df = df.drop(columns=['ORDEM','IDENT','CLASSE','CODIGO', 'CNPJ', 'ID_ECONO', 'NAICS_II', 'trimestre'])
df['DATA'] = pd.to_datetime(df['DATA'], format='%d%b%Y')


df['ANO'] = df['DATA'].dt.year
df['TRIMESTRE'] = df['DATA'].dt.quarter

df = df[~((df['ANO'] == 2009) & (df['TRIMESTRE'].isin([3, 4])))]
df.drop(columns=['ANO', 'TRIMESTRE'], inplace=True)
df = df.sort_values(by=['NOME', 'DATA']).reset_index(drop=True)
df = df[~df['NAICS_I'].str.contains('Servi.os financeiros e seguros', case=False, na=False, regex=True)]
df = df.drop(columns=['NAICS_I'])
df
```

```python
df = df.dropna(subset=['FCFF', 'FCFE', 'g_FCFF', 'g_FCFE', 'lnAT', 'ROIC', 'b_NOPAT', 'GIRO_AT', 'LEV', 'LC', 'ROE', 'b_PL'])
df = df[df['PL'] > 0]
df
```

```python
df.to_csv('clean_panel.csv')
```
