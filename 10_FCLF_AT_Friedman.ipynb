```python
import pandas as pd
import numpy as np

df_knn = pd.read_csv('KNN_resultados_FCLF_AT.csv')
df_xgb = pd.read_csv('XGBoost_resultados_FCLF_AT.csv')
df_ARIMA = pd.read_csv('ARIMA_resultados_FCLF_AT.csv')
df_RF = pd.read_csv('RF_resultados_FCLF_AT.csv')
df_Lasso = pd.read_csv('Lasso_resultados_FCLF_AT.csv')
```

```python
df_knn = df_knn.drop(columns=["Unnamed: 0", "Overfitting?", "Underfitting?"])
df_knn = df_knn.rename(columns={
    'Real':'Valores Reais',
    'Previsto':'knn_Previsto',
    'Erro Absoluto':'knn_MAE',
    'Train RMSE':'knn_train_RMSE',
    'Test RMSE':'knn_test_RMSE'
        })
```

```python
df_xgb = df_xgb.drop(columns=["Unnamed: 0", "Overfitting?", "Underfitting?"])
df_xgb = df_xgb.rename(columns={
    'Real':'Valores Reais',
    'Previsto':'xgb_Previsto',
    'Erro Absoluto':'xgb_MAE',
    'Train RMSE':'xgb_train_RMSE',
    'Test RMSE':'xgb_test_RMSE'
        })
```

```python
df_ARIMA = df_ARIMA.drop(columns=["Unnamed: 0", "Overfitting?", "Underfitting?"])
df_ARIMA = df_ARIMA.rename(columns={
    'Real':'Valores Reais',
    'Previsto':'ARIMA_Previsto',
    'Erro Absoluto':'ARIMA_MAE',
    'Train RMSE':'ARIMA_train_RMSE',
    'Test RMSE':'ARIMA_test_RMSE'
        })
```

```python
df_RF = df_RF.drop(columns=["Unnamed: 0", "Overfitting?", "Underfitting?"])
df_RF = df_RF.rename(columns={
    'Real':'Valores Reais',
    'Previsto':'RF_Previsto',
    'Erro Absoluto':'RF_MAE',
    'Train RMSE':'RF_train_RMSE',
    'Test RMSE':'RF_test_RMSE'
        })
```

```python
df_Lasso = df_Lasso.drop(columns=["Unnamed: 0", "Overfitting?", "Underfitting?"])
df_Lasso = df_Lasso.rename(columns={
    'Real':'Valores Reais',
    'Previsto':'Lasso_Previsto',
    'Erro Absoluto':'Lasso_MAE',
    'Train RMSE':'Lasso_train_RMSE',
    'Test RMSE':'Lasso_test_RMSE'
        })
```

```python
df = pd.merge(df_knn, df_xgb, on=['Empresa', 'Data', 'Valores Reais'], how='right')
df = pd.merge(df, df_ARIMA, on=['Empresa', 'Data', 'Valores Reais'], how='right')
df = pd.merge(df, df_RF, on=['Empresa', 'Data', 'Valores Reais'], how='right')
df = pd.merge(df, df_Lasso, on=['Empresa', 'Data', 'Valores Reais'], how='right')
df = df.dropna()
df
```

```python
from sklearn.metrics import mean_squared_error
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Função para calcular RMSE
def calcular_rmse_knn(df):
    return np.sqrt(mean_squared_error(df['Valores Reais'], df['knn_Previsto']))

# Aplicar por grupo (por empresa)
rmse_por_empresa_knn = df.groupby('Empresa').apply(calcular_rmse_knn).reset_index()
rmse_por_empresa_knn.columns = ['Empresa', 'RMSE_knn']
```

```python
# Função para calcular RMSE
def calcular_rmse_xgb(df):
    return np.sqrt(mean_squared_error(df['Valores Reais'], df['xgb_Previsto']))

# Aplicar por grupo (por empresa)
rmse_por_empresa_xgb = df.groupby('Empresa').apply(calcular_rmse_xgb).reset_index()
rmse_por_empresa_xgb.columns = ['Empresa', 'RMSE_xgb']
```

```python
# Função para calcular RMSE
def calcular_rmse_ARIMA(df):
    return np.sqrt(mean_squared_error(df['Valores Reais'], df['ARIMA_Previsto']))

# Aplicar por grupo (por empresa)
rmse_por_empresa_ARIMA = df.groupby('Empresa').apply(calcular_rmse_ARIMA).reset_index()
rmse_por_empresa_ARIMA.columns = ['Empresa', 'RMSE_ARIMA']
```

```python
def calcular_rmse_RF(df):
    return np.sqrt(mean_squared_error(df['Valores Reais'], df['RF_Previsto']))

# Aplicar por grupo (por empresa)
rmse_por_empresa_RF = df.groupby('Empresa').apply(calcular_rmse_RF).reset_index()
rmse_por_empresa_RF.columns = ['Empresa', 'RMSE_RF']
```

```python
def calcular_rmse_Lasso(df):
    return np.sqrt(mean_squared_error(df['Valores Reais'], df['Lasso_Previsto']))

# Aplicar por grupo (por empresa)
rmse_por_empresa_Lasso = df.groupby('Empresa').apply(calcular_rmse_Lasso).reset_index()
rmse_por_empresa_Lasso.columns = ['Empresa', 'RMSE_Lasso']
```

```python
rmse_final = pd.merge(rmse_por_empresa_knn, rmse_por_empresa_xgb, on=['Empresa'], how='right')
rmse_final = pd.merge(rmse_final, rmse_por_empresa_ARIMA, on=['Empresa'], how='right')
rmse_final = pd.merge(rmse_final, rmse_por_empresa_RF, on=['Empresa'], how='right')
rmse_final = pd.merge(rmse_final, rmse_por_empresa_Lasso, on=['Empresa'], how='right')

empresa_maior_rmse_knn = rmse_final.loc[rmse_final['RMSE_knn'].idxmax()]
empresa_menor_rmse_knn = rmse_final.loc[rmse_final['RMSE_knn'].idxmin()]

# Maior e menor RMSE para o XGB
empresa_maior_rmse_xgb = rmse_final.loc[rmse_final['RMSE_xgb'].idxmax()]
empresa_menor_rmse_xgb = rmse_final.loc[rmse_final['RMSE_xgb'].idxmin()]

# Maior e menor RMSE para o ARIMA
empresa_maior_rmse_ARIMA = rmse_final.loc[rmse_final['RMSE_ARIMA'].idxmax()]
empresa_menor_rmse_ARIMA = rmse_final.loc[rmse_final['RMSE_ARIMA'].idxmin()]

# Maior e menor RMSE para o RF
empresa_maior_rmse_RF = rmse_final.loc[rmse_final['RMSE_RF'].idxmax()]
empresa_menor_rmse_RF = rmse_final.loc[rmse_final['RMSE_RF'].idxmin()]

# Maior e menor RMSE para o Lasso
empresa_maior_rmse_Lasso = rmse_final.loc[rmse_final['RMSE_Lasso'].idxmax()]
empresa_menor_rmse_Lasso = rmse_final.loc[rmse_final['RMSE_Lasso'].idxmin()]

print("Maior RMSE KNN:", empresa_maior_rmse_knn)
print("Menor RMSE KNN:", empresa_menor_rmse_knn)

print("Maior RMSE XGB:", empresa_maior_rmse_xgb)
print("Menor RMSE XGB:", empresa_menor_rmse_xgb)

print("Maior RMSE ARIMA:", empresa_maior_rmse_ARIMA)
print("Menor RMSE ARIMA:", empresa_menor_rmse_ARIMA)

print("Maior RMSE RF:", empresa_maior_rmse_RF)
print("Menor RMSE RF:", empresa_menor_rmse_RF)

print("Maior RMSE Lasso:", empresa_maior_rmse_Lasso)
print("Menor RMSE Lasso:", empresa_menor_rmse_Lasso)
```

```python
from scipy.stats import friedmanchisquare
from scipy.stats import rankdata
import scikit_posthocs as sp

friedman_MAE = friedmanchisquare(grupo['knn_MAE'], grupo['xgb_MAE'], grupo['ARIMA_MAE'], grupo['RF_MAE'], grupo['Lasso_MAE']) 

friedman_MAE
```

```python
metric_cols = ['knn_MAE', 'xgb_MAE', 'ARIMA_MAE', 'RF_MAE', 'Lasso_MAE']   
model_names = ['KNN', 'XGB', 'ARIMA', 'RF', 'Lasso']

df_sp_MAE = grupo[metric_cols]
df_sp_MAE.columns = model_names

df_sp_MAE_long = df_sp_MAE.melt(var_name='Modelo', value_name='MAE')

dunn_MAE = sp.posthoc_dunn(df_sp_MAE_long, group_col='Modelo', val_col='MAE', p_adjust='bonferroni')
dunn_MAE
```

```python
# Aplicar ranking por linha (menor MAE = melhor)
ranks_MAE = grupo[['knn_MAE', 'xgb_MAE', 'ARIMA_MAE', 'RF_MAE', 'Lasso_MAE']].apply(rankdata, axis=1)
mean_ranks_MAE = ranks_MAE.mean(axis=0)

# Exibir ranking médio
print("Rank médio da MAE dos modelos:")
for i, rank in enumerate(mean_ranks_MAE, start=1):
    print(f"Modelo {i}: {rank:.3f}")
```

```python
friedman_RMSE_train = friedmanchisquare(grupo['knn_train_RMSE'], grupo['xgb_train_RMSE'], grupo['ARIMA_train_RMSE'], grupo['RF_train_RMSE'], grupo['Lasso_train_RMSE']) 

friedman_RMSE_train
```

```python
metric_cols = ['knn_train_RMSE', 'xgb_train_RMSE', 'ARIMA_train_RMSE', 'RF_train_RMSE', 'Lasso_train_RMSE']   
model_names = ['KNN', 'XGB', 'ARIMA', 'RF', 'Lasso']

df_sp_RMSE = grupo[metric_cols]
df_sp_RMSE.columns = model_names

df_sp_RMSE_long = df_sp_RMSE.melt(var_name='Modelo', value_name='RMSE')

dunn_RMSE = sp.posthoc_dunn(df_sp_RMSE_long, group_col='Modelo', val_col='RMSE', p_adjust='bonferroni')
dunn_RMSE
```

```python
# Aplicar ranking por linha (menor MAE = melhor)r
ranks_RMSE_train = grupo[['knn_train_RMSE', 'xgb_train_RMSE', 'ARIMA_train_RMSE', 'RF_train_RMSE', 'Lasso_train_RMSE']].apply(rankdata, axis=1)
mean_ranks_RMSE_train = ranks_RMSE_train.mean(axis=0)

# Exibir ranking médio
print("Rank médio da RMSE do treino dos modelos:")
for i, rank in enumerate(mean_ranks_RMSE_train, start=1):
    print(f"Modelo {i}: {rank:.3f}")
```
